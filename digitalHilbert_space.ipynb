```python
def generate_sparse_matrix(m, n, density):
    dense_matrix = np.random.randint(0, 11, size=(m, n))
    non_zero_mask = np.random.rand(m, n) < density
    sparse_matrix = dense_matrix * non_zero_mask
    
    return np.array(sparse_matrix)
```

## Looking at one cavity

```python
def cavity_hamiltonian(omega, N):
    a = destroy(N)
    # Planck constant throughout set to hbar=1
    return omega*a.dag()*a
```

```python
f_a = 1               # frequency (Hz)
omega_a = 2*pi*f_a    # angular frequency (rad/s)
kappa_a = 0.05        # cavity dissipation rate (i.e. loss rate)
N_a = 10

H_a = cavity_hamiltonian(omega_a, N_a)

H_a
```

### Cavity evolution (only loss)

```python
psi0 = basis(N_a, N_a-int(N_a/2))
tlist = np.linspace(0, 2*1/kappa_a, 101)
a = destroy(N_a)
c_ops = [sqrt(kappa_a)*a]

output = mesolve(H_a, psi0, tlist, c_ops, [a.dag() * a])

# Here H is the system Hamiltonian, psi0 is the initial state, tlist being the range of temporal evolution
# c_ops are the Linbladian operators, and we require the output to be the eigenvalues of the a^dagger a operator
# i.e. in this case the number of excitation in the cavity.
```

```python
fig, (ax) = plt.subplots(ncols=1, figsize=(5, 4))

ax.plot(tlist, output.expect[0], color='red',linewidth=3.5)
ax.set_xlim([min(tlist), max(tlist)])
ax.tick_params(axis='both', direction='in', which='both', labelsize=15)
ax.tick_params(axis='both', which='major', length=10, width=2.0)
plt.xlabel('Time evolution (s)', fontsize=15)
plt.ylabel(r'$\langle \hat{n}_a \rangle$', fontsize=15)
```

### Visualization of the Wigner function

To calculate the Wigner function in QuTiP, we first re-calculate the evolution without specifying any expectation value operators, which will result in that the solver return a list of density matrices for the system for the given time coordinates.

```python
density_mat = mesolve(H_a, psi0, tlist, c_ops, [])
```

```python
t_idx = np.where([tlist == t for t in [0.0, 20.0, 40.0]])[1]
tlist[t_idx]
```

```python
rho_list = np.array(density_mat.states)[t_idx]
```

```python
xvec = np.linspace(-5, 5, 200)
```

```python
fig, axes = plt.subplots(ncols=len(rho_list), sharex=True, figsize=(3*len(rho_list),3))

for idx, rho in enumerate(rho_list):
    rho = Qobj(rho)
    W = wigner(rho, xvec, xvec)
    axes[idx].contourf(xvec, xvec, W, 100, norm=mcolors.Normalize(-.25,.25), cmap=plt.get_cmap('RdBu'))
    axes[idx].set_title(r"$t = %.1f$" % tlist[t_idx][idx], fontsize=15)
    axes[idx].tick_params(axis='both', direction='in', which='both', labelsize=15)
    axes[idx].tick_params(axis='both', which='major', length=10, width=1.0)
    axes[idx].tick_params(axis='both', which='minor', length=7, width=1.0)
    axes[idx].xaxis.set_major_locator(MultipleLocator(5))
    axes[idx].yaxis.set_major_locator(MultipleLocator(3))
    
fig.tight_layout()
```

## Looking at one cavity coupled to one transmon

The system Hamiltonian reads

\begin{align}
H_{JC} = \omega_c \hat{a}^\dagger\hat{a} + \dfrac{1}{2}\omega_q \hat{\sigma}_z + g(\hat{a}^\dagger+\hat{a})(\hat{\sigma}_+ + \hat{\sigma}_-)
\end{align}

This JC model can yield qualitative agreement with experiments in situations where only the first two levels of the transmon (e.g $|\sigma=g, e\rangle$) play an important role. It is often the case, however, that quantitative agreement between theoretical predictions and experiments is obtained only when accounting for higher transmon energy levels and the multimode nature of the field.

Using the RWA (eliminating terms that don't conserve energy), we arrive at

\begin{align}
H_{JC}^{RWA} = \omega_c \hat{a}^\dagger\hat{a} + \dfrac{1}{2}\omega_q \hat{\sigma}_z + g(\hat{a}^\dagger\hat{\sigma}_- + \hat{a}\hat{\sigma}_+)
\end{align}

```python
f_c = 6                 # GHz 
omega_c = 2*pi*f_c
f_q = 4                 # GHz
omega_q = 2*pi*f_q
# If f_c = f_q we'll have transmon-cavity detuning = 0
g = 0.05 * 2 * pi       # coupling strength
N = 50                  # number of Fock states

tlist = np.linspace(0, 250, 1000)
```

```python
a_c = tensor(destroy(N), qeye(2))
sm = tensor(qeye(N), destroy(2))

H_RWA_JC = omega_c*a_c.dag()*a_c + omega_q*sm.dag()*sm + g*(a_c.dag()*sm + a_c*sm.dag())
```

```python
c_ops = [tensor(qeye(N), qeye(2))]
```

```python
psi0 = tensor(basis(N, 0), basis(2, 1))
```

```python
output = mesolve(H_RWA_JC, psi0, tlist, c_ops, [a_c.dag() * a_c, sm.dag() * sm])
```

```python
n_c = output.expect[0]
n_a = output.expect[1]

fig, axes = plt.subplots(1, 1, figsize=(7,4))

axes.plot(tlist, n_c, label="Cavity")
axes.plot(tlist, n_a, label="Atom excited state")
axes.legend(loc=0)
axes.set_xlabel('Time')
axes.set_ylabel('Occupation probability')
axes.set_title('Vacuum Rabi oscillations')
```

## Looking at two cavities coupled to one transmon

```python
f_a = 6                  # GHz 
omega_a = 2*pi*f_a
f_b = 7                  # GHz
omega_b = 2*pi*f_b
f_c = 4                  # GHz between the two
omega_c = 2*pi*f_c       # This is the plasma (ang.) frequency, not the transition frequency bt ground and first excited state. Scale as sqrt(EJEC)   
alpha = 0.3              # GHz transmon anharmonicity
omega_10 = omega_c - alpha
f_d = 4.1
omega_d = 2*pi*f_d

Na = 9                  # number of Fock states
Nb = 9                  # number of Fock states
Nc = 4                  # number of Fock states (assuming--actually this is true regardless--
Nd = 4                  # the transmon states are Fock states)


EJEC = 50                # transmon regime EJ/EC >> 1
e = 1.602e-19            # electron charge

V_a = 1                  # strengths of the effective voltage fluctuation due to E-fields
V_b = 1
g_a = -sqrt(2)*1j*e*V_a*(8*EJEC)**(-1/4)
g_b = -sqrt(2)*1j*e*V_b*(8*EJEC)**(-1/4)

Omega_d = 0.1            # 100 MHz
```

```python
g_a 
```

```python
g_b
```

Entering the rotating frame with $\omega_d$

```python
def delta(x):
    if x == 'da':
        return omega_d - omega_a
    if x == 'db':
        return omega_d - omega_b
    if x == 'dc':
        return omega_d - omega_c 
    if x == 'a':
        return omega_a - omega_10
    if x == 'b':
        return omega_b - omega_10
    if x == 'd':
        return omega_d - omega_10
```

### Defining operators

```python
a = tensor(destroy(Na), qeye(Nb), qeye(Nc))
b = tensor(qeye(Na), destroy(Nb), qeye(Nc))
c = tensor(qeye(Na), qeye(Nb), destroy(Nc))
```

### Hamiltonian generation

```python
H_I_RWA = (g_a * a + g_b * b) * c.dag() + (conj(g_a) * a.dag() + conj(g_b) * b.dag()) * c

H_RWA = - delta('da') * a.dag() * a - delta('db') * b.dag() * b \
        - delta('dc') * c.dag() * c - (alpha/2) * (c.dag() * c + 1) * c.dag() * c \
        + Omega_d * c.dag() + conj(Omega_d) * c + H_I_RWA
```

This Hamiltonian is controlled by the set of six dimensionless parameters,

\begin{align}
g_{a(b)}/\delta_{a(b)},\quad \delta_{a(b)}/\alpha,\quad \Omega_d/\delta_d,\quad \delta_d/\alpha
\end{align}

```python
eigenstates_qutip = H_RWA.eigenstates()[1]
eigenvalues_qutip = H_RWA.eigenstates()[0]
```

### Finding eigenvectors using CSR's eigsh

```python
H_RWA_csr = sparse_matrix_to_csr(H_RWA)
```
