```python
def EJEC(omega, alpha):
    '''
    Calculate EJEC ratio.
    Note that omega is in rad/s and anharmonicity alpha is positive
    '''
    f01 = omega/(2*pi)
    f12 = f01 - alpha
    EC = h*(alpha)
    EJ = (h*f01+EC)**2/(8*EC)
    
    return EJ/EC
```

# Fitting for 4 mode systems

## The system Hamiltonian

We're looking at two cavities, two transmons. One out of the two transmons is coupled to both cavities, while the other one is coupled to only one. The system Hamiltonian reads

\begin{align}
    H = H_{cav} + H_{anc}(t) + H_I
\end{align}

where the cavities are in their single-mode approximation,

\begin{align}
    H_{cav} = \hbar\omega_a a^\dagger a + \hbar\omega_b b^\dagger b,
\end{align}

and the ancillary components are the typical superconducting transmon circuits,

\begin{align}
    H_{anc}(t) = \sum_{j=1}^2 4E_{C_j} n_j^2 - E_{J_j}\cos(\phi_j) - 2e n_j V_{d_j}\sin(\omega_{d_j} t + \theta_{d_j}),
\end{align}

and finally the interaction terms,

\begin{align}
    H_I = -2e n_1 [V_a(a^\dagger + a) + V_b(b^\dagger +b)] - 2e n_2 V_a(a^\dagger + a),
\end{align}

where $V_a,V_b$ are the strengths of effective voltage fluctuations due to electric fields from the cavity modes $a,b$.

### Rotating wave approximation

We switch to a frame that rotates at the drive frequency $\omega_d$ by making a unitary $U=\exp(-i(a^\dagger a +b^\dagger b + c^\dagger c + e^\dagger e)\omega_d t)$. 

\begin{align}
    H_{RWA} &= -\hbar\delta_{d_1a}a^\dagger a - \hbar\delta_{d_1b}b^\dagger b+H^{RWA}_{anc}+H_I^{RWA},\\
    H^{RWA}_{anc}/\hbar &= -\delta_{d_1c}c^\dagger c - \dfrac{\alpha}{2}(c^\dagger c + 1)c^\dagger c + \Omega_{d_1} c^\dagger + \Omega_{d_1}^* c -\delta_{d_2e} e^\dagger e - \dfrac{\beta}{2}(e^\dagger e + 1)e^\dagger e+\Omega_{d_2} e^\dagger + \Omega_{d_2}^* e,\\
    H^{RWA}_I/\hbar &= (g_{ac} a + g_{bc} b)c^\dagger + (g_{ac}^* a^\dagger + g_{bc}^* b^\dagger) c + (g_{ae} a + g_{be} b)e^\dagger + (g_{ae}^* a^\dagger + g_{be}^* b^\dagger) e. 
\end{align}

The Hamiltonian is now controlled by a total of 10 dimensionless parameters. Specifically, the static part of the system is controlled by three sets of parameters,

\begin{align}
r_1 = \dfrac{\delta_{a,c10}}{\alpha},\\
r_2 = \dfrac{\delta_{b,c10}}{\alpha},\\
r_3 = \dfrac{\delta_{a,e10}}{\beta},\\
r_4 = \dfrac{g_{ac}}{\delta_{a,c10}} \ll 1,\\
r_5 = \dfrac{g_{bc}}{\delta_{b,c10}} \ll 1,\\
r_6 = \dfrac{g_{ae}}{\delta_{a,e10}} \ll 1.
\end{align}

Suppose that we can drive both of the transmons at the same time, the driving part is controlled by 

\begin{align}
r_7 = \dfrac{\delta_{d_1}}{\alpha},\\
r_8 = \dfrac{\Omega_{d_1}}{\delta_{d_1}} = \dfrac{\Omega_{d_1}}{r_7\alpha}\\
r_9 = \dfrac{\delta_{d_2}}{\beta},\\
r_{10} = \dfrac{\Omega_{d_2}}{\delta_{d_2}} = \dfrac{\Omega_{d_2}}{r_8\beta}
\end{align}

```python
def detuning_set(omega_a, omega_b, omega_c10, omega_e10):
    
    delta_ac = (omega_a - omega_c10)/(2*pi)
    delta_bc = (omega_b - omega_c10)/(2*pi)
    delta_ae = (omega_a - omega_e10)/(2*pi)
    
    return delta_ac, delta_bc, delta_ae
```

```python
def coupling_strength_set(r4, r5, r6):
    
    g_ac = delta_ac * r4
    g_bc = delta_bc * r5
    g_ae = delta_ae * r6
    
    return g_ac, g_bc, g_ae
```

```python
def drive_parameter_set(r7, r8, r9, r10):
    
    delta_d1 = r7*alpha # GHz
    omega_d1 = omega_c10 + 2*pi*delta_d1
    Omega_d1 = r8*delta_d1
    
    delta_d2 = r9*beta # GHz
    omega_d2 = omega_e10 + 2*pi*delta_d2 
    Omega_d2 = r10*delta_d2 
    
    delta_d1a = (omega_d1 - omega_a)/(2*pi) 
    delta_d1b = (omega_d1 - omega_b)/(2*pi) 
    delta_d1c = (omega_d1 - omega_c)/(2*pi) 
    delta_d2e = (omega_d2 - omega_e)/(2*pi)
    
    return delta_d1a, delta_d1b, delta_d1c, delta_d2e, Omega_d1, Omega_d2
```

## Hamiltonian generation

```python
def H_RWA(r4, r5, r6, r7, r8, r9, r10, couplingOn, driveOn):
    
    '''
        Hamiltonian generation
    '''
    switch1 = 0
    switch2 = 0
    delta_d1d2 = 0
    
    if couplingOn == True:
        switch1 = 1
    if driveOn == True:
        switch2 = 1
       
    g_ac, g_bc, g_ae = coupling_strength_set(r4, r5, r6)
    delta_d1a, delta_d1b, delta_d1c, delta_d2e, Omega_d1, Omega_d2 = drive_parameter_set(r7, r8, r9, r10) 
        
    H_ANC_RWA = - delta_d1c * c.dag() * c - (alpha/2) * (c.dag() * c + 1) * c.dag() * c \
                - delta_d2e * e.dag() * e - (beta/2)  * (e.dag() * e + 1) * e.dag() * e
    
    H_DRV_RWA = Omega_d1 * c.dag() + conj(Omega_d1) * c + Omega_d2 * e.dag() + conj(Omega_d2) * e
    
    H_INT_RWA = (g_ac * a + g_bc * b) * c.dag() + (conj(g_ac) * a.dag() +conj(g_bc) * b.dag()) * c \
                + g_ae * a * e.dag() + conj(g_ae) * a.dag() * e
    
    H_TOT_RWA = - delta_d1a * a.dag() * a - delta_d1b * b.dag() * b \
                + H_ANC_RWA + switch1 * H_INT_RWA + switch2 * H_DRV_RWA
    
    return H_TOT_RWA
```
